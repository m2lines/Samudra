```python
import xarray as xr
```

```python
data = xr.open_zarr("/mnt/home/sd5313/data/Ocean/3D_data")
```

```python
# Print range of years
print(data.isel(time=0)["time"].values, data.isel(time=3999)["time"].values)

# Difference in years
(data.isel(time=3999)["time"].values - data.isel(time=0)["time"].values).days / 365.25
```

```python
import math

tenyear_samples = math.ceil(365.25 * 10 // 5)
tenyear_data = data.isel(time=slice(4000 - tenyear_samples, None))
print(
    tenyear_data.isel(time=0)["time"].values,
    tenyear_data.isel(time=tenyear_samples)["time"].values,
)

# Difference in years
(
    tenyear_data.isel(time=tenyear_samples)["time"].values
    - tenyear_data.isel(time=0)["time"].values
).days / 365.25
```

```python
# Save the 10 year data
tenyear_data.to_zarr("/mnt/home/sd5313/data/Ocean/3D_data_10year")
```

```python
from dask.diagnostics import ProgressBar

with ProgressBar():
    tenyear_data_mean = tenyear_data.mean().compute()
    tenyear_data_std = tenyear_data.std().compute()

tenyear_data_mean.to_zarr("/mnt/home/sd5313/data/Ocean/3D_data_10year_means")
tenyear_data_std.to_zarr("/mnt/home/sd5313/data/Ocean/3D_data_10year_stds")
```
