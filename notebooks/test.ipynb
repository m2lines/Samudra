```python
mean_path = "/mnt/home/sd5313/data/2024-11-12-cm4-piControl-ocean-200yr-dataset-stats/centering.nc"
std_path = "/mnt/home/sd5313/data/2024-11-12-cm4-piControl-ocean-200yr-dataset-stats/scaling-full-field.nc"

mean_ds = xr.open_dataset(mean_path)
std_ds = xr.open_dataset(std_path)

mean_ds
std_ds
```

```python
my_mean = xr.open_dataset("/mnt/home/sd5313/data/CM4_5daily_v0.4.0_means.zarr")
my_std = xr.open_dataset("/mnt/home/sd5313/data/CM4_5daily_v0.4.0_stds.zarr")

my_mean
my_std
```

```python
depth_str_toind_map = {
    "2_5": 0,
    "10_0": 1,
    "22_5": 2,
    "40_0": 3,
    "65_0": 4,
    "105_0": 5,
    "165_0": 6,
    "250_0": 7,
    "375_0": 8,
    "550_0": 9,
    "775_0": 10,
    "1050_0": 11,
    "1400_0": 12,
    "1850_0": 13,
    "2400_0": 14,
    "3100_0": 15,
    "4000_0": 16,
    "5000_0": 17,
    "6000_0": 18,
}
```

```python
# Compare mean and std
vars = [
    var
    for var in my_mean.variables
    if var
    not in [
        "DSWRFtoa",
        "air_temperature_at_two_meters",
        "hfds_anomalies",
        "surface_temperature",
        "lat_b",
        "lon_b",
    ]
]
ai2_vars = []
for var in vars:
    if "lev" in var:
        var, depth_str = var.split("_lev_")
        ai2_vars.append(var + "_" + str(depth_str_toind_map[depth_str]))
    else:
        ai2_vars.append(var)
```

```python
ai2_vars
```

```python
import numpy as np

np.allclose(
    mean_ds[ai2_vars].to_array().values, my_mean[vars].to_array().values, atol=1e-1
)
np.allclose(
    std_ds[ai2_vars].to_array().values, my_std[vars].to_array().values, atol=1e-1
)
```
